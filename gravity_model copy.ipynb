```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
df = pd.read_excel("gravmodel.xlsx")

columns = []
for col in df.columns:
    columns.append(col)
print(columns)
```

```python
y = df['ln_tradevalue'] # dependent variable
X = df[['ln_distance', 'common_language', 'population_ue', 'population_usa', 'ln_gdp_ue', 'ln_gdp_usa', 'gdp_per_capita_ue', 'gdp_per_capita_usa', 'gatt_ue', 'gatt_usa', 'wto_ue', 'wto_usa', 'rta', 'entry_cost_ue', 'entry_cost_usa', 'entry_proc_ue', 'entry_proc_usa', 'entry_time_ue', 'entry_time_usa', 'entry_tp_ue', 'entry_tp_usa', 'AD_initations_ue', 'AD_measures_ue', 'AD_initations_usa', 'AD_measures_usa']]
```

```python
X = sm.add_constant(X) # adding a constant
lm = sm.OLS(y, X).fit() # fitting the model
lm.summary()
```

```python
df2 = pd.read_excel("gravity model.xlsx")
```

```python
df2['tradevalue'] = np.log(df2['tradevalue'])
df2[['distance', 'gdp_ue', 'gdp_usa']] = np.log(df2[['distance', 'gdp_ue', 'gdp_usa']])
df2
```

```python
y2 = df2['tradevalue'] # dependent variable
X2 = df2[['distance', 'common_language', 'legal_new_ue', 'legal_new_usa', 'comleg_pretrans', 'comleg_posttrans', 'heg_ue', 'col_dep_ever', 'population_ue', 'population_usa', 'gdp_ue', 'gdp_usa', 'gdp_per_capita_ue', 'gdp_per_capita_usa', 'gatt_ue', 'gatt_usa', 'wto_ue', 'wto_usa', 'rta', 'entry_cost_ue', 'entry_cost_usa', 'entry_proc_ue', 'entry_proc_usa', 'entry_time_ue', 'entry_time_usa', 'entry_tp_ue', 'entry_tp_usa', 'AD_initations_ue', 'AD_measures_ue', 'AD_initations_usa', 'AD_measures_usa']] # independent variable
```

```python
X2 = sm.add_constant(X2) # adding a constant
lm = sm.OLS(y2, X2).fit() # fitting the model
results = lm.summary(alpha=0.05)
results
```
